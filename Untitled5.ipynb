```python
import pandas as pd
import numpy as np


data = np.arange(1, 36).reshape(7, 5)

df = pd.DataFrame(
    data,
    columns=['A', 'B', 'C', 'D', 'E']
)

print(df)
```

```python
df_add = df + 10
print(df_add)
```

```python
df_sub = df - 5
print(df_sub)
```

```python
df_mul = df * 2
print(df_mul)
```

```python
df_div = df / 5
print(df_div)
```
